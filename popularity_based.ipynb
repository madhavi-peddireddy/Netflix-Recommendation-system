# Criteria For Population Based Recommendation system

The criteria is based on:
1. Movies with highest rating
2. Number of views

<b>Average ratings of movies

```python
movie_data.groupby('title')['rating'].mean().head()
```

```python
movie_data.groupby('title')['rating'].mean().sort_values(ascending=False).head(50)
```

```python
movie_data
```

```python
movie_data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
rating_means_count=pd.DataFrame(movie_data.groupby('title')['rating'].mean())
```

```python
rating_means_count
```

```python
rating_means_count['rating_Count']=pd.DataFrame(movie_data.groupby('title')['rating'].count())
```

```python
rating_means_count
```

```python
rating_means_count['rating']=round(rating_means_count['rating'],1)
```

```python
rating_means_count
```

```python
rating_means_count=rating_means_count[(rating_means_count['rating']>4) & (rating_means_count['rating_Count']>100)]
```

```python
rating_means_count
```

```python
rating_means_count=rating_means_count.sort_values(by='rating', ascending=False)
```
